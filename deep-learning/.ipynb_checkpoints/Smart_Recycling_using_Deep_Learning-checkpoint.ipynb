## Smart Recycling using Deep Learning

Every day, we put things into our recycle bin, to reduce landfill waste. As every city council has slightly different rules for recycling, I am attempting to build a technological solution to ensure you only recycle things that are permitted by your local council. More discussions about recycling contamination can be found [here](https://www.cleanup.org.au/recycle).

### Problem Definition

<ul>
<li>An efficient and functional recycling process is crucial for societal sustainability and the environment. Such a recycling process involves collecting and processing materials that can be turned into new products.</li>

<li>Unfortunately due to our lack of awareness and understanding about the recycling rules and recycling process, often times we hinder the recycling process by comingling the non-recyclable items along with the recyclables. For example, throwing any single-use plastic bag with a hard plastic container jeopardizes the recycling of the hard plastic container. Similarly, textile waste causes the waste sorting machines to malfunction. The major contaminators are textiles, food and single-use plastic bags.</li> 

<li><p>So, comingled wastes and putting wrong items in wrong bin can cause recycling contamination that might ruin the possibility to recover the recyclable items.</li>

<li>However, <b>there is a lack of mechanism for visually identifying a piece of waste produced at the source (in our households or business)</b> which clearly indicates whether the piece of waste is recyclable or not.</li>

```python
def make_confusion_matrix(cfm):
    off_diag_mask = np.eye(*cfm.shape, dtype=bool)
    vmin = np.min(cfm)
    vmax = np.max(cfm)

    #Plotting the confusion matrix
    plt.figure(figsize=(6,5))
    sns.heatmap(cfm, annot=True, fmt='g', mask=~off_diag_mask, cmap='Blues', vmin=vmin, vmax=vmax)
    sns.heatmap(cfm, annot=True, fmt='g', mask=off_diag_mask, vmin=vmin, vmax=vmax)

    plt.title('Predicted Values')
    plt.ylabel('Actal Values')

    plt.show()
```

### Resize the image (224 x 224)

```python
def image_resizer(image, width=224, height=224):
    (img_height, img_width) = image.shape[:2]
    d_width = 0
    d_height = 0
    
    if img_width < img_height:
        image = resize(image, width=width, inter=cv2.INTER_AREA)
        d_height = int((image.shape[0] - height) / 2.0)
    else:
        image = resize(image, height=height, inter=cv2.INTER_AREA)
        d_width = int((image.shape[1] - width) / 2.0)

    (res_height, res_width) = image.shape[:2]
    image = image[d_height:res_height - d_height, d_width:res_width - d_width]
    return cv2.resize(image, (width, height), interpolation=cv2.INTER_AREA)
```

### Converting the image to array of floats to make an input of the Convnet model

```python
def image_to_array(image):
    return img_to_array(image)
```

### Image Preprocess

```python
def preprocess(image, image_height, image_width):
    image = image_resizer(image, image_height, image_width)
    image = image_to_array(image)
    return image
```

### Loading image from disk and preprocessing and transforming image

```python
def image_loader(dir_name, image_height, image_width):
    image_paths = list(paths.list_images(dir_name))
    img_data = []
    img_labels = []
    for image_path in image_paths:
        image = cv2.imread(image_path)
        label = image_path.split(os.path.sep)[-2]
        image = preprocess(image, image_height, image_width)
        img_data.append(image)
        img_labels.append(label)
    return (np.array(img_data), np.array(img_labels))
```

```python
def image_loader_64(dir_name):
    image_paths = list(paths.list_images(dir_name))
    img_data = []
    img_labels = []
    for image_path in image_paths:
        image = cv2.imread(image_path)
        label = image_path.split(os.path.sep)[-2]
        image = preprocess(image, 64, 64)
        img_data.append(image)
        img_labels.append(label)
    return (np.array(img_data), np.array(img_labels))
```

## Building Model 1 - Building Resnet From Scratch

### Model Configuration

```python
bn_epsilon = 0.00002
bn_momentum = 0.9
reg = 0.0005
classes = 7
batch_size = 64
```

### Building Model -1 Resnet from scratch - Stages = (3, 3, 3), Filters = (64, 64, 128, 256)

```python
def build_resnet_block_one(x):
    b1_batch_norm_1 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(x)
    b1_activation_1 = Activation("relu")(b1_batch_norm_1)
    b1_conv_1 = Conv2D(16, (1, 1), use_bias=False, kernel_regularizer=l2(reg))(b1_activation_1)

    b1_batch_norm_2 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(b1_conv_1)
    b1_activation_2 = Activation("relu")(b1_batch_norm_2)
    b1_conv_2 = Conv2D(16, (3, 3), strides=(1,1), padding="same", 
                          use_bias=False, kernel_regularizer=l2(reg))(b1_activation_2)

    b1_batch_norm_3 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(b1_conv_2)
    b1_activation_3 = Activation("relu")(b1_batch_norm_3)
    b1_conv_3 = Conv2D(64, (1, 1), use_bias=False, kernel_regularizer=l2(reg))(b1_activation_3)

    b1_shortcut = Conv2D(64, (1, 1), strides=(1,1), use_bias=False, kernel_regularizer=l2(reg))(b1_activation_1)
    x = add([b1_conv_3, b1_shortcut])


    b2_shortcut = x
    b2_batch_norm_1 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(x)
    b2_activation_1 = Activation("relu")(b2_batch_norm_1)
    b2_conv_1 = Conv2D(16, (1, 1), use_bias=False, kernel_regularizer=l2(reg))(b2_activation_1)

    b2_batch_norm_2 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(b2_conv_1)
    b2_activation_2 = Activation("relu")(b2_batch_norm_2)
    b2_conv_2 = Conv2D(16, (3, 3), strides=(1,1), 
                          padding="same", use_bias=False, kernel_regularizer=l2(reg))(b2_activation_2)

    b2_batch_norm_3 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(b2_conv_2)
    b2_activation_3 = Activation("relu")(b2_batch_norm_3)
    b2_conv_3 = Conv2D(64, (1, 1), use_bias=False, kernel_regularizer=l2(reg))(b2_activation_3)

    x = add([b2_conv_3, b2_shortcut])

    b3_shortcut = x
    b3_batch_norm_1 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(x)
    b3_activation_1 = Activation("relu")(b3_batch_norm_1)
    b3_conv_1 = Conv2D(16, (1, 1), use_bias=False, kernel_regularizer=l2(reg))(b3_activation_1)

    b3_batch_norm_2 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(b3_conv_1)
    b3_activation_2 = Activation("relu")(b3_batch_norm_2)
    b3_conv_2 = Conv2D(16, (3, 3), strides=(1,1), 
                          padding="same", use_bias=False, kernel_regularizer=l2(reg))(b3_activation_2)

    b3_batch_norm_3 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(b3_conv_2)
    b3_activation_3 = Activation("relu")(b3_batch_norm_3)
    b3_conv_3 = Conv2D(64, (1, 1), use_bias=False, kernel_regularizer=l2(reg))(b3_activation_3)

    x = add([b3_conv_3, b3_shortcut])
    return x
    
def build_resnet_block_two(x):
    b1_batch_norm_1 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(x)
    b1_activation_1 = Activation("relu")(b1_batch_norm_1)
    b1_conv_1 = Conv2D(32, (1, 1), use_bias=False, kernel_regularizer=l2(reg))(b1_activation_1)

    b1_batch_norm_2 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(b1_conv_1)
    b1_activation_2 = Activation("relu")(b1_batch_norm_2)
    b1_conv_2 = Conv2D(32, (3, 3), strides=(2,2), padding="same", 
                          use_bias=False, kernel_regularizer=l2(reg))(b1_activation_2)

    b1_batch_norm_3 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(b1_conv_2)
    b1_activation_3 = Activation("relu")(b1_batch_norm_3)
    b1_conv_3 = Conv2D(128, (1, 1), use_bias=False, kernel_regularizer=l2(reg))(b1_activation_3)

    b1_shortcut = Conv2D(128, (1, 1), strides=(2,2), use_bias=False, kernel_regularizer=l2(reg))(b1_activation_1)
    x = add([b1_conv_3, b1_shortcut])

    b2_shortcut = x
    b2_batch_norm_1 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(x)
    b2_activation_1 = Activation("relu")(b2_batch_norm_1)
    b2_conv_1 = Conv2D(32, (1, 1), use_bias=False, kernel_regularizer=l2(reg))(b2_activation_1)

    b2_batch_norm_2 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(b2_conv_1)
    b2_activation_2 = Activation("relu")(b2_batch_norm_2)
    b2_conv_2 = Conv2D(32, (3, 3), strides=(1,1), 
                          padding="same", use_bias=False, kernel_regularizer=l2(reg))(b2_activation_2)

    b2_batch_norm_3 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(b2_conv_2)
    b2_activation_3 = Activation("relu")(b2_batch_norm_3)
    b2_conv_3 = Conv2D(128, (1, 1), use_bias=False, kernel_regularizer=l2(reg))(b2_activation_3)

    x = add([b2_conv_3, b2_shortcut])

    b3_shortcut = x
    b3_batch_norm_1 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(x)
    b3_activation_1 = Activation("relu")(b3_batch_norm_1)
    b3_conv_1 = Conv2D(32, (1, 1), use_bias=False, kernel_regularizer=l2(reg))(b3_activation_1)

    b3_batch_norm_2 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(b3_conv_1)
    b3_activation_2 = Activation("relu")(b3_batch_norm_2)
    b3_conv_2 = Conv2D(32, (3, 3), strides=(1,1), 
                          padding="same", use_bias=False, kernel_regularizer=l2(reg))(b3_activation_2)

    b3_batch_norm_3 = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(b3_conv_2)
    b3_activation_3 = Activation("relu")(b3_batch_norm_3)
    b3_conv_3 = Conv2D(128, (1, 1), use_bias=False, kernel_regularizer=l2(reg))(b3_activation_3)

    x = add([b3_conv_3, b3_shortcut])
    return x
    
```

```python
def build_resnet():
    # apply BN => ACT => POOL
    inputs = Input(shape=(input_shape_64))
    x = BatchNormalization(axis=channel_dim, epsilon=bn_epsilon, momentum=bn_momentum)(inputs)
    x = Conv2D(64, (3, 3), use_bias=False, padding="same", kernel_regularizer=l2(reg))(x)


    x = build_resnet_block_one(x)
    x = build_resnet_block_two(x)

    x = BatchNormalization(axis=-1, epsilon=bn_epsilon, momentum=bn_momentum)(x)
    x = Activation("relu")(x)

    x = AveragePooling2D((8, 8))(x)
    x = Flatten()(x)
    x = Dense(classes, kernel_regularizer=l2(reg))(x)
    x = Activation("softmax")(x)

    model = Model(inputs, x, name="resnet") 
    return model
```

## Building Model 2 : Employing Transfer Learning by building Model 2 and using pre-trained weights

```python
from tensorflow.keras.applications import ResNet50

class PretrainedResnet50:
    @staticmethod
    def buildHead(baseModel, classes, D):
        headModel = baseModel.output
        headModel = Flatten(name="flatten")(headModel)
        headModel = Dense(D, activation="relu")(headModel)
        headModel = Dropout(0.5)(headModel)

        headModel = Dense(7, activation="softmax")(headModel)
        return headModel
    
    @staticmethod
    def build():
        baseModel = ResNet50(include_top=False, weights="imagenet", pooling='avg', input_tensor=Input(shape=(224, 224, 3)))
        headModel = PretrainedResnet50.buildHead(baseModel, 7, 256)
        model = Model(inputs=baseModel.input, outputs=headModel)

        for layer in baseModel.layers:
            layer.trainable = False
        return model
```

### Generate adversarial image by introducing perturbations - Using fast gradient sign method

```python
def generate_adversary(model, image, label, eps=2 / 255.0):
    image = tf.cast(image, tf.float32)

    with tf.GradientTape() as tape:
        tape.watch(image)
        pred = model(image)
        loss = tf.keras.losses.CategoricalCrossentropy(label, pred)

    gradient = tape.gradient(loss, image)
    signedGrad = tf.sign(gradient)

    adversary = (image + (signedGrad * eps)).numpy()
    return adversary
```

```python
def generate_adversarial_image(model, image, label, eps=2 / 255.0):
    image = tf.cast(image, tf.float32)

    with tf.GradientTape() as tape:
        tape.watch(image)
        pred = model(image)
        loss = tf.keras.losses.MSE(label, pred)

    gradient = tape.gradient(loss, image)
    signedGrad = tf.sign(gradient)

    adversary = (image + (signedGrad * eps)).numpy()
    return adversary
```

```python
def generate_adversarial_batch(model, total, images, labels, dims, eps=0.01):
    (height, width, depth) = dims

    while True:
        perturb_imgs = []
        perturb_lbls = []
        
        ## working on random indexes
        idxs = np.random.choice(range(0, len(images)), size=total, replace=False)
        for i in idxs:
            image = images[i]
            label = labels[i]
            adv_img = generate_adversarial_image(model, image.reshape(1, height, width, depth), label, eps=eps)

            perturb_imgs.append(adv_img.reshape(height, width, depth))
            perturb_lbls.append(label)

        yield (np.array(perturb_imgs), np.array(perturb_lbls))
```
