## Pre-processing steps 

<p> As pre-processing step I have applied Vectorization and Value Normalization to the data.</p>

<p> Since neural networks optimizes differentiable continuous functions so the networks expect floating point tensors as inputs as continuous values. The process of converting the input data into a tensors is called Vectorization. As the datatype of the train and test datasets are uint8, I have cast the training and test images to float32 tensors. </p>

```python
train_X_vect = train_X_full.reshape((train_X_full.shape[0], 28 * 28))
print("Training Data after reshaping = ", train_X_vect.shape)

test_X_vect = test_X.reshape((test_X.shape[0], 28 * 28))
print("Test Data after reshaping     = ", test_X_vect.shape)
```

```python
train_X_float = train_X_vect.astype('float32')
test_X_float = test_X_vect.astype('float32')
```

<p> I have also applied normalization on the tensor of float32 data. This will turn the data into values in the range of 0 to 1. This will help the optimization algorithm like Gradient Descent to converge faster. If the inputs are in a different scales the error function can be elongated and the optimization process becomes slower. This takes a lot of time to converge.</p>

    <span style="color: #0000FF; font-weight:bold">As part of Pre-Processing Step I have done the following:</span><br> 
<ol>
    <li><b>Vectorization</b> - I reshaped the 28 X 28 pixel values to a Vector of dimension 784 X 1 (784-dimension Vector) to be used as an input to the Neural network for all the 60000 training instances and 10000 test instances.<br><br><b>Reason for performing Vectorization</b> is that Neural Networks work on input vectors that is fed into the next dense layer, after initializing with weights and biases.<br><br></li>
    <li><b>Type Conversion</b> - I converted(casted) the data type of Pixels from uint8 (Unsigned Integer with values between 0-255 as shown above) to float32 (Floating Point).<br><br><b>Reason for casting from uint8 to float32</b> is that Neural Networks work on continuous values both for inputs and outputs of the Network.<br><br></li>
    <li><b>Normalization</b> - I rescaled the pixel values that are in the range of 0-255 as can be seen from the pixels values of the image above to a range of floating point numbers in the range of 0-1.<br><br><b>Reason for Normalizing the inputs</b> is that:
        <ul>
            <li>There is a big disparity in the pixel values in some images as can be seen by the pixel values printed above with some being in high 200s while others being less than 40 and Normalization will reduce this disparity between pixel values.<br><br></li>
            <li>Also, Normalization will reduce the much large pixel values, like pixel values that are greater than 100, sometimes reaching close to 255, as compared to the initial weight values of the Neural Network.
            <br><br>
            </li>
            <li>The above steps would have the combined effect of helping the Neural network to converge faster.
            </li>
        </ul>
    </li>
</ol>

## Setting up model

### Experiment 1 

<p> Constucting a deep learning model having only fully connected (dense) layer.</p>

```python
model = Sequential()
model.add(Dense(1024, activation='relu', input_shape=(784,)))
model.add(Dense(512, activation='relu', kernel_initializer="he_normal"))
model.add(Dense(256, activation='relu', kernel_initializer="he_normal"))
model.add(Dense(128, activation='relu', kernel_initializer="he_normal"))
model.add(Dense(64, activation='relu', kernel_initializer="he_normal"))
model.add(Dense(32, activation='relu', kernel_initializer="he_normal"))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

<ul><b>Output layer:</b> 
    <li><b># of output nodes</b></li>
    <li>I have used 10 output nodes in the last dense layer. Since this is a Multiclass image classification problem and we have 10 categories, the dimension of the output layer would be 10. Hence in the model architecture the last Dense layer dimension has been defined as 10<br><br></li>
    <li>In the output layer I have used softmax activation function </li>
</ul>


<ul><b>Hidden layer:</b> 
    <li>I have constructed deep neural network architecture. The network has 6 hidden layers<br><br></li>
    <li>
        layer 1 ------ 1024<br>
        layer 2 ------ 512<br>
        layer 3 ------ 256<br>
        layer 4 ------ 128<br>
        layer 5 ------ 64<br>
        layer 6 ------ 32
    </li>
    <li>
        <br>
        layer 1 ------ ReLU<br>
        layer 2 ------ ReLU<br>
        layer 3 ------ ReLU<br>
        layer 4 ------ ReLU<br>
        layer 5 ------ ReLU<br>
        layer 6 ------ ReLU
    </li>
</ul>


<ul><b>Input layer:</b> 
    <li>As the image has 28 x 28 pixels hence the input size is 28 x 28 x 1</li>
    <li><p>
        The model takes 1 dimensional column vector of shape 784 X 1 tensor. Hence the input 28 x 28 has been  reshaped to 784 x 1 column vector to fit the input dimension of the first Dense layer</p></li>
    
    
<ul>
    <p> My design decisions </p><br>
        <ul>
            <li><b>The input layer</b></li>
            <li>The input layer shape here is 784, the actual input has been reshaped to fit the input dimension</li>
            <li><b>The Network architecture</b></li>
            <li>The network has 6 layers, each having 1024, 512, 256, 128, 32 and 10 layer. I wanted to ensure that the model gets high dimensional representation space while learning complex patterns and also does not overfit too much. I know I can regularize the network using different measures if necessary. So I first tried to create a deep network with suffient number of neurons to learn as much as possible.</li>
            <li><b>The Activation functions for Intermediate layers</b></li>
            <li><p>The hidden (intermediate) layers are all going through the ReLU activation function, in order to transform the affine value in a non-liner representation. ReLU function is continuous but not differentiable at 0, and it has proven to have worked well in the Neural Network training</li>
            <li><b>The output layer</b></li>
            <li>As we have a 10 class muticlass classification problem the output layer has 10 classes each correspondin to one output category</li>
            <li><b>The output activation function</b></li>
            <li>I am using Softmax activation function for the final output value. Softmax ensures that the estimated probabilities are between 0 to 1</li>
        </ul> 
    
</ul>

## Fitting the model

### Initializing TensorBoard

```python
root_logdir = os.path.join(os.curdir, "nn_training_logs")

def get_log_location():
    run_id = time.strftime("run_%Y_%m_%d-%H_%M_%S")
    return os.path.join(root_logdir, run_id)

log_dir = get_log_location()
log_dir
```

### Compiling the model

```python
adam_optimizer = keras.optimizers.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999)
```

```python
model.compile(loss='categorical_crossentropy', optimizer=adam_optimizer, metrics=['accuracy'])
```

### Creating the validation set for model training with categorical labels to use with cross_entropy loss

```python
valid_X, train_X = train_X_norm[:5000], train_X_norm[5000:]
valid_y_cat, train_y_cat = train_y_full_cat[:5000], train_y_full_cat[5000:]
```

### Creating the validation labels for model training with non- categorical labels to use with sparse cross_entropy loss

```python
valid_y, train_y = train_y_full[:5000], train_y_full[5000:]
```

### Configurations

```python
n_epochs = 30
n_batch_size = 128
```

### Fitting the model

```python
from keras import callbacks

tensorboard_cb = keras.callbacks.TensorBoard(log_dir)

history = model.fit(train_X, train_y_cat, batch_size= n_batch_size, epochs=n_epochs,
                    validation_data=(valid_X, valid_y_cat), callbacks =[tensorboard_cb])
```

```python
test_loss, test_acc = model.evaluate(test_X_norm, test_y_cat)

print('Experiment 1:')
print(f'test_loss: {test_loss}')
print(f'test_acc: {test_acc}')
print()
```
